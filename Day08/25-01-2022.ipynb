**Categorical Plot**
- sns.catplot()

```python
help(sns.catplot)
```

```python
sns.catplot(x = 'species', y = 'sepal_length', data = df)
```

```python
sns.catplot(x = 'species', y = 'sepal_length', data = df, kind = 'box')
```

```python
sns.catplot(x = 'species', y = 'sepal_length', data = df, kind = 'box', palette = 'Green')
```

```python
sns.catplot(x = 'species', y = 'sepal_length', data = df, 
           kind = 'box', palette = 'gnuplot2_r')
```

```python
sns.catplot(x = 'species', y = 'sepal_length', data = df, 
           kind = 'violin', palette = 'YlOrRd')
```

**Joint Plot**
- Combination of two plots
    - sns.jointplot()

```python
df.head(3)
```

```python
sns.jointplot(x = 'sepal_length', y = 'petal_length', data = df, color = 'orange')
```

```python
sns.jointplot(x = 'species', y = 'petal_length', data = df, color = 'orange')
```

```python
help(sns.jointplot)
```

```python
sns.jointplot(x = 'sepal_length', y = 'petal_length', data = df, color = 'orange', kind = 'hex')
```

```python
## KDE: Kernel Density Estimation
#### Probability density curve
sns.jointplot(x = 'sepal_length', y = 'petal_length', data = df, color = 'm', kind = 'kde')
```

**Pair Plot**
- Pairwise relationship between each and every column
    - sns.pairplot()

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'species')
```

```python
sns.pairplot(df, hue = 'species', palette = 'husl', markers = ['o','d','*'])
```

**Heatmaps**

```python
t.corr()
```

```python
sns.heatmap(t.corr(), annot = True)
```
