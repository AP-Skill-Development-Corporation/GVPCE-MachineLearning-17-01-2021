### Logistic Regression
- It is a classification model
- If target column contains binary values(like 0 or 1, yes or no, happy or sad), then logistic regression is the best classification model to apply.
- In this model, we are applying Sigmoid curve formula.



- ***Here y is linear regression (y = b0 + b1 * x)***

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/Tshirt_size.csv')
df
```

```python
X = df[['Height', 'Weight']] # Independent or input
Y = df['Shirt_Size']         # dependent or target
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(X, Y) ## training
```

```python
Y_predict = log.predict(X)  ## testing
Y_predict
```

```python
# give height = 150, weight = 52
# what is predicted shirt size?
```

```python
log.predict([[150, 52]])
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(Y, Y_predict)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y, Y_predict)
```

### Support Vector Machine
- It is a Linera Classifier
- It is used to find the best hyperplane.
- Hyperplane is a line which divides the dataset into 2 classifier.


**Maximum Width**
- w * x - b = +1
