### Combining
- join 2 data sets
- Adding a new row
- Adding a new column

```python
dict1 = {'Games':['Carams', 'Ludo', 'PubG'], 'Score': [9, 8.7, 9.4]}
d1 = pd.DataFrame(dict1)
d1
```

```python
dict2 = {'Games':['TempleRun','Candy Crush','Zomby'], 'Score': [9.2, 9.6, 7]}
d2 = pd.DataFrame(dict2)
d2
```

**Joining 2 datasets**
- pd.concat([d1, d2])

```python
pd.concat([d1, d2])
```

- Axis = 0: Indicates rows
- Axis = 1: Indictaes columns

```python
pd.concat([d1, d2], axis = 1)
```
